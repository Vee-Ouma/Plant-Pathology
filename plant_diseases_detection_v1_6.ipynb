```
"""
Created on Thu Oct 31 15:25:38 2019

@author: Fakrul Islam
Feature Extraction Tomato Early Blight
"""


from pathlib import Path
import numpy as np
import joblib
from keras.preprocessing import image
from keras.applications import vgg16

# Path to folders with tomato early blight training data
tomato_earlyblight_path = Path("/content/drive/My Drive/plant/tomato/early_blight/training_data/early_blight_yes")

# Path to not tomato early blight folders 
not_earlyblight_path = Path("/content/drive/My Drive/plant/tomato/early_blight/training_data/early_blight_no")


images = []
labels = []


# Load all not-tomato earlyblight images
for img in not_earlyblight_path.glob("*.png"):
    # Load the image from disk
    img = image.load_img(img)

    # Convert the image to a numpy array
    image_array = image.img_to_array(img)

    # Add the image to the list of images
    images.append(image_array)

    # For each 'not early blight' image, the expected value should be 0
    labels.append(0)

# Load all the tomato earlyblight images
for img in tomato_earlyblight_path.glob("*.png"):
    # Load the image from disk
    img = image.load_img(img)

    # Convert the image to a numpy array
    image_array = image.img_to_array(img)

    # Add the image to the list of images
    images.append(image_array)

    # For each 'early blight' image, the expected value should be 1
    labels.append(1)

        
# Create a single numpy array with all the images we loaded
x_train = np.array(images)

# Also convert the labels to a numpy array
y_train = np.array(labels)

# Normalize image data to 0-to-1 range
x_train = vgg16.preprocess_input(x_train)

# Load a pre-trained neural network to use as a feature extractor
pretrained_nn = vgg16.VGG16(weights='imagenet', include_top=False, input_shape=(64, 64, 3))

# Extract features for each image (all in one pass)
features_x = pretrained_nn.predict(x_train)

# Save the array of extracted features to a file
joblib.dump(features_x, "/content/drive/My Drive/plant/tomato/early_blight/x_train_tomato_earlyblight.dat")

# Save the matching array of expected values to a file
joblib.dump(y_train, "/content/drive/My Drive/plant/tomato/early_blight/y_train_tomato_earlyblight.dat")
```

```
"""
Created on Mon Oct 31 15:29:07 2019

@author: Fakrul Islam
Train Tomato Early Blight

"""

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from pathlib import Path
import joblib
import matplotlib.pyplot as plt
import numpy


print("Train Tomato Early Blight")
# Load data set
x_train = joblib.load("/content/drive/My Drive/plant/tomato/early_blight/x_train_tomato_earlyblight.dat")
y_train = joblib.load("/content/drive/My Drive/plant/tomato/early_blight/y_train_tomato_earlyblight.dat")

# Create a model and add layers
model = Sequential()

model.add(Flatten(input_shape=x_train.shape[1:]))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=['accuracy']
)

# Train the model
history=model.fit(
    x_train,
    y_train,
    epochs=100,
    shuffle=True
)

#print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.title('early blight model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.title('early blight model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Save neural network structure
model_structure = model.to_json()
f = Path("/content/drive/My Drive/plant/tomato/early_blight/model_structure_tomato_earlyblight.json")
f.write_text(model_structure)

# Save neural network's trained weights
model.save_weights("/content/drive/My Drive/plant/tomato/early_blight/model_weights_tomato_earlyblight.h5")
```

```
"""
Created on Mon Nov  4 09:11:29 2019
Bulk Data Validation for Early Blight
@author: Fakrul Islam
"""

from keras.models import model_from_json
from pathlib import Path
from keras.preprocessing import image
import numpy as np
from keras.applications import vgg16

validation_image_path = Path("/content/drive/My Drive/plant/tomato/early_blight/validation")
early_blight_json_path = "/content/drive/My Drive/plant/tomato/early_blight/model_structure_tomato_earlyblight.json"
early_blight_weight_path = "/content/drive/My Drive/plant/tomato/early_blight/model_weights_tomato_earlyblight.h5"


#search Pattern for Early Blight
def searchPatternEarlyBlight(imagePath):            
    f = Path(early_blight_json_path)
    model_structure = f.read_text()
    model = model_from_json(model_structure)
    model.load_weights(early_blight_weight_path)
    img = image.load_img(imagePath, target_size=(64, 64))
    image_array = image.img_to_array(img)
    images = np.expand_dims(image_array, axis=0)
    images = vgg16.preprocess_input(images)
    feature_extraction_model = vgg16.VGG16(weights='imagenet', include_top=False, input_shape=(64, 64, 3))
    features = feature_extraction_model.predict(images)
    results = model.predict(features)
    single_result = results[0][0]
    print ("Search Pattern in Early Blight\n")
    print(imagePath)
    print("Likelihood that this image contains a traget image: {}%".format(int(single_result * 100)))


if __name__ == "__main__":
   print ("Start Image Pattern Searching")  
   for imagePath in validation_image_path.glob("*.png"):
       #print(imagePath) 
       searchPatternEarlyBlight(imagePath)

   print("End Image Pattern Searching")
```
